```python
import json 
```

```python
#Read the json file .
f=open("Record.json","r")
r=f.read()
f.close()
```

```python
#Convert the (str) type of json file into dictionary using loads function.
records=json.loads(r)
records
```

#Add more items in the list of items.

```python
#Here, Adding multiple items in the Record.json at one time.
for i in range(0,7):
    Prod_id=input("Enter the id of a product:")
    name=input("Name of a product:")
    Unit_Price=input("Enter the price of a product:")
    Qty=input("Enter the number of a product:")
    Manufacturing_date=input("Enter the manufactured date of a product:")
    Expiring_date=input("Enter the expired date of a product:")
    records[Prod_id]={'name':name,'Unit_Price':Unit_Price,'Qty': Qty,'Manufacturing_date':Manufacturing_date,'Expiring_date':Expiring_date}
    js=json.dumps(records)
    fd=open("Record.json","w")
    fd.write(js)
    fd.close()
```
